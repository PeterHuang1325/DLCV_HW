```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import os
```

```python
path = '../hw4_data/office/'
train_df = pd.read_csv(os.path.join(path, 'train.csv'))
val_df =  pd.read_csv(os.path.join(path, 'val.csv'))
val_df
```

```python
labelencoder = LabelEncoder()
train_df['enc_lbl'] = labelencoder.fit_transform(train_df['label'])
val_df['enc_lbl'] = labelencoder.fit_transform(val_df['label'])
train_df
```

```python
enc2class = dict()
encodings = list(train_df['enc_lbl'])
class_labels = list(train_df['label'])
for i, enc in enumerate(encodings):
    enc2class[enc] = class_labels[i]
```

```python
enc2class
```
